## 📌 Step 1: Import NumPy

### Why?
- Used for matrix operations.
- Required for implementing Multi-Head Attention.

### Remember
- Multi-Head Attention is built using matrix multiplication.

```python
import numpy as np 
print("Numpy Version:",np.__version__)
```

## 📌 Step 2: Create Embedding Matrix

### Why?
- Convert words into numerical vectors.
- Every attention head receives the same embeddings.

### Remember
- Input is shared by all attention heads.

```python
X = np.array([
    [1.0, 0.0],
    [0.5, 1.0],
    [1.0, 1.0]
])

print("Embedding Matrix")
print(X)
print("\nShape:", X.shape)
```

## 📌 Step 3: Create Weights for Head 1

### Why?
- Each head has its own Wq, Wk and Wv.
- These weights learn different relationships.

### Remember
- Different heads = Different weight matrices.

```python
Wq1 = np.array([
    [2,1],
    [1,2]
])

Wk1 = np.array([
    [1,0],
    [0,1]
])

Wv1 = np.array([
    [1,2],
    [2,1]
])
```

## 📌 Step 4: Create Weights for Head 2

### Why?
- Head 2 should learn a different representation.
- Therefore it uses different trainable weights.

### Remember
- Head 1 and Head 2 never share weights.

```python
Wq2 = np.array([
    [1,2],
    [2,1]
])

Wk2 = np.array([
    [2,0],
    [1,1]
])

Wv2 = np.array([
    [2,1],
    [1,2]
])
```

## 📌 Step 5: Compute Q, K and V for Head 1

### Formula

Q₁ = X × Wq₁

K₁ = X × Wk₁

V₁ = X × Wv₁

### Remember
- Head 1 processes the sentence independently.

```python
Q1 = X @ Wq1
K1 = X @ Wk1
V1 = X @ Wv1
```

## 📌 Step 6: Compute Q, K and V for Head 2

### Formula

Q₂ = X × Wq₂

K₂ = X × Wk₂

V₂ = X × Wv₂

### Remember
- Same input.
- Different learned representation.

```python
Q2 = X @ Wq2
K2 = X @ Wk2
V2 = X @ Wv2
```

```python
scores1 = Q1 @ K1.T
scores1 = scores1 / np.sqrt(K1.shape[1])

def softmax(x):
    exp = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp / np.sum(exp, axis=-1, keepdims=True)

attention1 = softmax(scores1)

output1 = attention1 @ V1

print(output1)
```

```python
scores2 = Q2 @ K2.T
scores2 = scores2 / np.sqrt(K2.shape[1])

attention2 = softmax(scores2)

output2 = attention2 @ V2

print(output2)
```

## 📌 Step 9: Concatenate Head Outputs

### Why?
- Combine information learned by different heads.

### Formula

Output = Concat(Head₁, Head₂)

### Remember
- Concatenation joins features.
- It does not add them.

```python
multi_head_output = np.concatenate((output1, output2), axis=1)

print(multi_head_output)

print("\nShape:", multi_head_output.shape)
```

```python
Wo = np.array([
    [1,0],
    [0,1],
    [1,1],
    [2,1]
])

final_output = multi_head_output @ Wo

print(final_output)
```
