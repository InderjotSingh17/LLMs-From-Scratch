## 📌 Step 1: Import NumPy

### Why?
- NumPy is used for matrix and vector operations.
- Transformers rely heavily on matrix multiplication.

### Output
- Displays the installed NumPy version.

```python
import numpy as np
print("Numpy Version:",np.__version__)
```

## 📌 Step 2: Create Embeddings

### Why?
- Computers understand numbers, not words.
- Every word is converted into an embedding vector.
- Here we manually create embeddings for learning.

### Remember
Rows → Words

Columns → Embedding Dimensions

```python
X = np.array([
    [1.0, 0.0],      # I
    [0.5, 1.0],      # love
    [1.0, 1.0]       # AI
])
print("Embedding Matrix:")
print(X)
print("\nShape:",X.shape)
```

## 📌 Step 3: Create Wq, Wk and Wv

### Why?
- Wq creates Queries.
- Wk creates Keys.
- Wv creates Values.
- These are trainable parameters.

### Formula
Q = X × Wq

K = X × Wk

V = X × Wv

```python
Wq = np.array([
    [2,1],
    [1,2]
])

Wk = np.array([
    [1,0],
    [0,1]
])

Wv = np.array([
    [1,2],
    [2,1]
])

print("Wq")
print(Wq)

print("\nWk")
print(Wk)

print("\nWv")
print(Wv)
```

## 📌 Step 4: Generate Query Matrix (Q)

### Why?
- Convert embeddings into Query vectors.
- Query represents what each word is searching for.

### Formula
Q = X × Wq

### Remember
- Q is **not** attention.
- Every word gets its own Query vector.

```python
Q = X @ Wq
print("Queries")
print(Q)
print("\nShape:", Q.shape)
```

## 📌 Step 5: Generate Key Matrix (K)

### Why?
- Convert embeddings into Key vectors.
- Keys represent the information each word contains.

### Formula
K = X × Wk

### Remember
- Every Query is compared with every Key.

```python
K = X @ Wk

print("Keys")
print(K)

print("\nShape:", K.shape)
```

## 📌 Step 6: Generate Value Matrix (V)

### Why?
- Convert embeddings into Value vectors.
- Values contain the information passed to the output.

### Formula
V = X × Wv

### Remember
- Attention decides **which Values** should contribute more.

```python
V = X @ Wv

print("Values")
print(V)

print("\nShape:", V.shape)
```

## 📌 Step 7: Compute Attention Scores

### Why?
- Measure similarity between Queries and Keys.

### Formula
Scores = Q × Kᵀ

### Remember
- Higher score → Higher similarity.
- Scores are **not probabilities**.

```python
scores = Q @ K.T

print("Attention Scores")
print(scores)

print("\nShape:", scores.shape)
```

## 📌 Step 8: Scale Attention Scores

### Why?
- Prevent very large attention scores.
- Helps Softmax produce stable probabilities.

### Formula
Scaled Scores = Scores / √dk

### Remember
dk = Key dimension

```python
dk = K.shape[1]

print("dk =", dk)

scaled_scores = scores / np.sqrt(dk)

print("\nScaled Scores")
print(scaled_scores)
```

## 📌 Step 9: Apply Softmax

### Why?
- Convert attention scores into probabilities.

### Remember
- Each row sums to 1.
- Higher probability → More attention.

```python
def softmax(x):
    exp = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp / np.sum(exp, axis=-1, keepdims=True)

attention = softmax(scaled_scores)

print("Attention Weights")
print(attention)
```

```python
print(attention.sum(axis=1))
```

## 📌 Step 10: Compute Final Output

### Why?
- Generate context-aware word representations.

### Formula
Output = Attention × V

### Remember
- Values are weighted by attention probabilities.
- This is the final output of Self-Attention.

```python
output = attention @ V

print("Final Output")
print(output)

print("\nShape:", output.shape)
```
