## 🎯 Objective

In this notebook, we'll understand why GPT uses a Key-Value (KV) Cache during text generation.

By the end of this notebook, you'll understand:

- What Keys and Values are
- Why recomputing attention is wasteful

# Imports

```python
import torch
```

# Sentence Being Generated

```python
sentence = [
    "I",
    "love",
    "learning",
    "AI"
]
print(sentence)
```

# Without KV Cache

```python
for step in range(1, len(sentence)+1):
    print(sentence[:step])
```

# Count the Work

```python
total_tokens = 0
for step in range(1, len(sentence)+1):
    total_tokens += step
print("Total Token Computations:", total_tokens)
```

# Phase 2 
KV CACHE CONCEPT

```python
import torch.nn.functional as F
# New token's Query
Q = torch.randn(1, 4)
# Cached Keys from previous tokens
K_cache = torch.randn(3, 4)
# Cached Values from previous tokens
V_cache = torch.randn(3, 4)
# Attention Scores
scores = Q @ K_cache.T
# Attention Weights
weights = F.softmax(scores, dim=-1)
# Final Output
output = weights @ V_cache
print(output)
```
