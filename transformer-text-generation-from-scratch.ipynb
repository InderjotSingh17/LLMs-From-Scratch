```python
import torch
import torch.nn.functional as F
```

```python
vocab = {
    0: "I",
    1: "love",
    2: "learning",
    3: "AI",
    4: ".",
    5: "Python",
    6: "because",
    7: "it",
    8: "is",
    9: "fun"
}
```

```python
logits = torch.tensor([
    2.1,
    0.8,
    4.5,
    3.2,
    0.5,
    2.7,
    1.0,
    0.4,
    1.6,
    2.2
])
print(logits)
```

```python
probabilities=F.softmax(logits,dim=0)
print(probabilities)
```

```python
for i,p in enumerate(probabilities):
    print(f"{vocab[i]} : {p.item():.3f}")
```

```python
next_token=torch.argmax(probabilities)
print("predicted token:",vocab[next_token.item()])
```

```python
temperature=0.7
scaled_logits=logits/temperature
probs=F.softmax(scaled_logits,dim=0)
print(probs)
```

```python
next_token = torch.multinomial(probs, num_samples=1)

print(vocab[next_token.item()])
```

```python
k = 3
topk_probs, topk_indices = torch.topk(probabilities, k)
topk_probs = topk_probs / topk_probs.sum()
sampled = torch.multinomial(topk_probs, 1)
next_token = topk_indices[sampled]
print(vocab[next_token.item()])
```

```python
sorted_probs, sorted_indices = torch.sort(probabilities, descending=True)

cumulative_probs = torch.cumsum(sorted_probs, dim=0)

p = 0.90

mask = cumulative_probs <= p

filtered_probs = sorted_probs[mask]
filtered_indices = sorted_indices[mask]

filtered_probs = filtered_probs / filtered_probs.sum()

sampled = torch.multinomial(filtered_probs, 1)

next_token = filtered_indices[sampled]

print(vocab[next_token.item()])
```

```python
generated = []

for _ in range(5):

    probs = F.softmax(logits, dim=0)

    token = torch.multinomial(probs, 1)

    generated.append(vocab[token.item()])

print("Generated Sentence:")
print(" ".join(generated))
```

```python
methods = {
    "Greedy": "Highest probability only",
    "Temperature": "Controls randomness",
    "Top-k": "Choose among top k tokens",
    "Top-p": "Choose from smallest set whose cumulative probability reaches p"
}

for method, description in methods.items():
    print(f"{method} -> {description}")
```
