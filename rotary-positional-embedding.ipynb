```python
query = torch.tensor([2.0, 1.0])
key = torch.tensor([1.5, 2.5])
print("Query:", query)
print("Key:", key)
```

```python
def rotate(vector):
    return torch.tensor([
        -vector[1],
         vector[0]
    ])
```

```python
rotated_query = rotate(query)
rotated_key = rotate(key)
print("Rotated Query:", rotated_query)
print("Rotated Key:", rotated_key)
```

```python
original_score = torch.dot(query, key)
rotated_score = torch.dot(rotated_query, rotated_key)
print("Original Attention Score:", original_score.item())
print("RoPE Attention Score:", rotated_score.item())
```

```python
position = 2
theta = position * (math.pi / 4)
print("Theta (radians):", theta)
print("Theta (degrees):", math.degrees(theta))
```

```python
rotation_matrix = torch.tensor([
    [math.cos(theta), -math.sin(theta)],
    [math.sin(theta),  math.cos(theta)]
])
print(rotation_matrix)
```

```python
vector = torch.tensor([2.0, 1.0])
rotated_vector = rotation_matrix @ vector
print("Original:", vector)
print("Rotated :", rotated_vector)
```
