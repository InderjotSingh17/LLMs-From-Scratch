# Embedding Matrix

```python
embedding_matrix = [
    [0.11, 0.42, -0.30, 0.25],
    [0.60, -0.20, 0.71, -0.40],
    [0.32, 0.81, -0.15, 0.53],
    [0.42, -0.15, 0.88, 0.31]
]

for row in embedding_matrix:
    print(row)
```

# Random Initialization

```python
import random
vocab_size = 4
embedding_dim = 5
embedding_matrix = []
for _ in range(vocab_size):
    row = []
    for _ in range(embedding_dim):
        row.append(round(random.uniform(-1, 1), 2))
    embedding_matrix.append(row)
for row in embedding_matrix:
    print(row)
```

# Assign Words to Rows

```python
vocab = {
    "I": 0,
    "love": 1,
    "learning": 2,
    "AI": 3
}
for word, idx in vocab.items():
    print(word, "->", embedding_matrix[idx])
```

# Build Our Own Embedding Layer

```python
class MyEmbedding:
    def __init__(self, embedding_matrix):
        self.embedding_matrix = embedding_matrix
    def forward(self, token_ids):
        embeddings = []
        for token in token_ids:
            embeddings.append(self.embedding_matrix[token])
        return embeddings
```

# Create Our Layer

```python
layer = MyEmbedding(embedding_matrix)
```

```python
token_ids = [0, 1, 2, 3]
output = layer.forward(token_ids)
for embedding in output:
    print(embedding)
```

# Cosine Similarity

# why compare embeddings?

```python
embedding_ai = [0.42, -0.15, 0.88, 0.31]
embedding_python = [0.39, -0.18, 0.91, 0.29]
embedding_car = [-0.72, 0.64, -0.11, 0.48]
print("AI      :", embedding_ai)
print("Python  :", embedding_python)
print("Car     :", embedding_car)
```

# convert to tensors 

```python
import torch
ai = torch.tensor(embedding_ai)
python = torch.tensor(embedding_python)
car = torch.tensor(embedding_car)
```

# cosine similarity 
(How similar the directions of two vectors are.)

```python
cosine = torch.nn.CosineSimilarity(dim=0)
```

```python
similarity=cosine(ai,python)
print(similarity.item())
```

```python
similarity = cosine(ai, car)
print("Similarity:", similarity.item())
```
