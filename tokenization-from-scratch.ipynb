```python
sentence="I love learning AI"
tokens=[]
current_word=""
for character in sentence:
    if character!=" ":
        current_word+=character
    else:
        tokens.append(current_word)
        current_word=""
tokens.append(current_word)
print(tokens)
```

```python
vocab = {}
index = 0
for token in tokens:
    if token not in vocab:
        vocab[token] = index
        index += 1
print(vocab)
```

```python
token_ids=[]
for token in tokens:
    token_ids.append(vocab[token])
print(token_ids)
```

```python
corpus = [
    "I love AI",
    "AI loves Python",
    "Python loves AI",
    "I love Machine Learning"
]
print(corpus)
```

```python
all_tokens = []

for sentence in corpus:

    current_word = ""

    for character in sentence:

        if character != " ":
            current_word += character

        else:
            all_tokens.append(current_word)
            current_word = ""

    all_tokens.append(current_word)

print(all_tokens)
```

# BPE(BYTE PAIR ENCODING)

```python
words = [
    "play",
    "playing",
    "played",
    "player"
]

print(words)
```

```python
corpus = [
    "low",
    "lower",
    "lowest",
    "newest",
    "widest"
]

print(corpus)
```

```python
split_words = []

for word in corpus:
    split_words.append(list(word))

print(split_words)
```

```python
pair_counter = Counter()

for word in split_words:

    for i in range(len(word)-1):

        pair = (word[i], word[i+1])

        pair_counter[pair] += 1

print(pair_counter)
```

```python
most_common_pair = pair_counter.most_common(1)

print(most_common_pair)
```

```python
def merge_pair(words, pair):

    merged_words = []

    first, second = pair

    merged_token = first + second

    for word in words:

        new_word = []

        i = 0

        while i < len(word):

            if (
                i < len(word)-1 and
                word[i] == first and
                word[i+1] == second
            ):

                new_word.append(merged_token)

                i += 2

            else:

                new_word.append(word[i])

                i += 1

        merged_words.append(new_word)

    return merged_words
```

```python
pair = most_common_pair[0][0]

split_words = merge_pair(split_words, pair)

print(split_words)
```

```python
pair_counter = Counter()

for word in split_words:

    for i in range(len(word)-1):

        pair_counter[(word[i], word[i+1])] += 1

print(pair_counter)
```

```python
most_common_pair = pair_counter.most_common(1)

print(most_common_pair)
```

```python
split_words = merge_pair(
    split_words,
    most_common_pair[0][0]
)

print(split_words)
```

```python
split_words = [
    list(word)
    for word in corpus
]

for iteration in range(5):

    pair_counter = Counter()

    for word in split_words:

        for i in range(len(word)-1):

            pair_counter[(word[i], word[i+1])] += 1

    best_pair = pair_counter.most_common(1)[0][0]

    print(f"Iteration {iteration+1}")

    print("Best Pair:", best_pair)

    split_words = merge_pair(
        split_words,
        best_pair
    )

    print(split_words)
    print()
```

```python
print("Final Learned Tokens:")

unique_tokens = set()

for word in split_words:

    for token in word:

        unique_tokens.add(token)

print(sorted(unique_tokens))
```

```python
print("Word Tokenization      -> OOV Problem")
print("Character Tokenization -> Too Many Tokens")
print("BPE                    -> Best of Both Worlds")
```
